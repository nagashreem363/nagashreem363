```python
!pip install seaborn
import matplotlib.pyplot as plt
import seaborn as sns

season_sales = data1.groupby('Season')['PRICE_CURRENT'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='PRICE_CURRENT', data=season_sales)
plt.title('Total Sales by Season')
plt.xlabel('Season')
plt.ylabel('Total Sales')
plt.show()

highest_sales_season = season_sales.loc[season_sales['PRICE_CURRENT'].idxmax()]
print(f"The season with the highest sales is: {highest_sales_season['Season']}")
print(f"Total sales in this season: {highest_sales_season['PRICE_CURRENT']:.2f}")
```

```python


# Group by 'CATEGORY' and sum 'PRICE_CURRENT' to get total sales per category
category_sales = data1.groupby('CATEGORY')['PRICE_CURRENT'].sum().reset_index()

# Sort in descending order and get the top category
highest_sales_category = category_sales.sort_values(by='PRICE_CURRENT', ascending=False).iloc[0]

print(f"The category with the highest sales is: {highest_sales_category['CATEGORY']}")
print(f"Total sales in this category: {highest_sales_category['PRICE_CURRENT']:.2f}")


top_n = 10
top_categories = category_sales.sort_values(by='PRICE_CURRENT', ascending=False).head(top_n)

plt.figure(figsize=(12, 7))
sns.barplot(x='PRICE_CURRENT', y='CATEGORY', data=top_categories, palette='viridis')
plt.title(f'Top {top_n} Categories by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.show()
```

```python


# Group by 'PRODUCT_NAME' and sum 'PRICE_CURRENT' to get total sales per product
product_sales = data1.groupby('PRODUCT_NAME')['PRICE_CURRENT'].sum().reset_index()

# Sort in descending order and get the top product
highest_sales_product = product_sales.sort_values(by='PRICE_CURRENT', ascending=False).iloc[0]

print(f"The product with the highest sales is: {highest_sales_product['PRODUCT_NAME']}")
print(f"Total sales for this product: {highest_sales_product['PRICE_CURRENT']:.2f}")

# Create a bar plot for the top products (optional, to visualize more than just the top)
# Let's visualize the top 10 products for clarity
top_n = 10
top_products = product_sales.sort_values(by='PRICE_CURRENT', ascending=False).head(top_n)

plt.figure(figsize=(12, 7))
sns.barplot(x='PRICE_CURRENT', y='PRODUCT_NAME', data=top_products, palette='magma')
plt.title(f'Top {top_n} Products by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```
